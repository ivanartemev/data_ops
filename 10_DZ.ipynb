# 3 Задание

```python
ratings = pd.read_csv('C:\\Users\\Пользователь\\Desktop\\важно\\ml-latest-small\\ml-latest-small\\ratings.csv')
ratings.head()
```

```python
f = ratings.filter(items = ['userId', 'rating', 'timestamp'])#.ratings.groupby(["userId"]).agg({"rating": "count"}).reset_index()
```

```python
g = f.groupby('userId').agg({"rating": "count"}).reset_index()
g.head()
```

```python
a = f.groupby('userId').max().reset_index()

a.head()
```

```python
a_max_timestamp=a[['userId','timestamp']]
```

```python
a_max_timestamp.rename(columns={'timestamp': 'timestamp_max'}, inplace=True)

a_max_timestamp.head()
```

```python
b = f.groupby('userId').min().reset_index()

b.head()
```

```python
b_min_timestamp=b[['userId','timestamp']]
```

```python
b_min_timestamp.rename(columns={'timestamp': 'timestamp_min'}, inplace=True)

b_min_timestamp.head()
```

```python
join1 = g.merge(a_max_timestamp, how='left')

join1.head(3)
```

```python
all_merge = join1.merge(b_min_timestamp, how='left')

all_merge.head(3)
```

```python
all_merge['avg'] = all_merge['timestamp_max'] - all_merge['timestamp_min']
all_merge.reset_index()

all_merge.head(3)
```

```python
all_merge_ = all_merge[all_merge['rating'] > 100]
all_merge_.head()
```
